The vgsale sourced from Kaggle contains information on video game sales. the data has 11 columns and 16598 rows and covers game from 1980 t0 2020. The dataset includes detailed sales figures for North America, Europe, Japan, and other regions, along with global sales.
which are the top publishers?
Which platform has the most sales?
How do sale vary across genre?

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Vilars/Data-201/refs/heads/main/vgsales.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Year'] = df['Year'].fillna(df['Year'].median())
df = df.dropna()
df.head(15)
```

```python
# pairplot
sns.pairplot(df)
plt.show()
```

```python
# Histogram of annual distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Year'], kde = True, bins = 50)
plt.title('Annual Distribution of Video Games')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Global sale by genre
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='Global_Sales', data=df, showfliers=False)
plt.title('Global_Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global_Sales')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Platform'], df['NA_Sales'])
plt.title('NA_Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('NA_Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 sale by platform
plt.figure(figsize=(10, 6))

top_10_platforms = df.groupby('Platform')['Global_Sales'].sum().nlargest(10)

plt.bar(top_10_platforms.index, top_10_platforms.values)
plt.title('Top 10 Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot of top 10 publisher
plt.figure(figsize=(10, 6))
top_10_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
plt.bar(top_10_publishers.index, top_10_publishers.values)
plt.title('Top 10 Publishers')
plt.xlabel('Publisher')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
```

```python
# NA_sale vs EU_sale
plt.figure(figsize=(10, 8))
sns.scatterplot(x='EU_Sales', y='NA_Sales', data = df,hue='Genre', alpha=0.5, s=50)
plt.title('NA_Sales vs EU_Sales')
plt.xlabel('EU_Sales')
plt.ylabel('NA_Sales')
plt.xlim(0, 30)
plt.ylim(0, 45)
plt.show()
```

```python
# Sale trend over time
plt.figure(figsize=(10, 6))
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
yearly_sales = df.groupby(df['Year'].dt.year)['Global_Sales'].sum()
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-')
plt.title('Sale Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Simulation of a nonparametric inference of the Median

# Select a quantitative variable, for example, Global_Sales
data = df['Global_Sales']

# Simulate a nonparametric inference of the Median
sample_size = int(0.1 * len(data))
num_simulations = 1000

medians = []

for _ in range(num_simulations):
    sample = np.random.choice(data, size=sample_size, replace=True)
    medians.append(np.median(sample))

# Calculate the median of the medians
median_of_medians = np.median(medians)

print(f"The nonparametric inference of the median for Global_Sales is approximately {median_of_medians:.2f}.")
```
